# SVR

# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read World bank CSV

```python
df = pd.read_csv('WorldBank.csv')
```

```python
df
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
data = df[df["Indicator Name"] == 'Population ages 15-64 (% of total population)']
data
```

# Divide the data into input and output

```python
x = np.arange(1960,2020).reshape(-1,1)
x
```

```python
y = data.values[0][4:-1].reshape(-1,1)
```

```python
y
```

# Plot scatter plot of Population ages 15-64 (% of total population)

```python
plt.scatter(x,y)
plt.show()
```

# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

```python
from sklearn.preprocessing import StandardScaler

x_sc = StandardScaler()
x_trans = x_sc.fit_transform(x)
x_trans
```

```python
y_sc = StandardScaler()
y_trans = y_sc.fit_transform(y)
y_trans
```

```python
print("Scaling Complete.")
```

# Fitting SVR on Linear Kernel on dataset

```python
from sklearn.svm import SVR, LinearSVR
```

```python
svr = SVR(kernel='linear')
svr.fit(x_trans, y_trans.ravel())
```

# Predict Values using Model

```python
# use inverse_transform
```

```python
x_pred = svr.predict(x_trans)
x_pred
```

```python
y_pred_lin = y_sc.inverse_transform(x_pred.reshape(-1,1))
y_pred_lin
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred_lin)
plt.show()
```

# Predict the X_scaled using 3 Kernel

```python
# above Steps Only - Kernel Change
```

```python
svr_poly = SVR(kernel='poly', degree=3)
svr_rbf = SVR(kernel='rbf')

svr_poly.fit(x_trans,y_trans.ravel())
svr_rbf.fit(x_trans,y_trans.ravel())
```

# Visualising the  results

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y,color='darkorange',label='Data')

plt.plot(x,y_pred_lin,color='navy',label='Linear Model')

y_predicted_poly = y_sc.inverse_transform(svr_poly.predict(x_trans).reshape(-1,1))
plt.plot(x,y_predicted_poly, color='cornflowerblue', label='Polynomial Model')

y_predicted_rbf = y_sc.inverse_transform(svr_rbf.predict(x_trans).reshape(-1,1))
plt.plot(x,y_predicted_rbf, color='green', label='RBF Model')

plt.xlabel('data')
plt.show()
```
